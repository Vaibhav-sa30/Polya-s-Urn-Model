#Polya's Urn Model
Polya's urn model is designed to show how the probabilities of drawing different colored balls from an urn can change over time as balls are drawn and added back to the urn. The model explores the concept of dependent events, where the outcome of one event affects the probabilities of future events.

In the context of Polya's urn model:

The probabilities do change after every draw. This is because, as you draw balls and add them back with additional balls of the same color, the composition of the urn changes, influencing the likelihood of drawing certain colors in subsequent draws.

However, the confusion might arise because the probabilities eventually reach an equilibrium. This means that after a certain number of draws, the probabilities stabilize and stop changing, even though they do change from step to step initially. The probabilities stabilize in a way that they become constant, and this is what mathematical induction proves.

The mathematical induction used to show that probabilities remain unchanged is not about every individual step, but about the long-term equilibrium state. The inductive proof shows that regardless of how many steps you've taken, the probabilities eventually converge to certain values and stay constant after a certain point. This is an important property of Polya's urn model that helps in understanding the behavior of the system over a large number of steps.

So, to summarize, Polya's urn model initially involves changing probabilities after each draw due to changing composition, but these probabilities eventually reach an equilibrium state where they remain constant. Mathematical induction helps establish the long-term stability of these probabilities.
